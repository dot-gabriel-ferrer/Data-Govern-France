```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from os import walk
```

```python
# get filenames with walk
f = []
for (dirpath, dirnames, filenames) in walk('collision_data/'):
    f.extend(filenames)
    break
```

```python
f.index
```

```python
from codecs import ignore_errors
from tkinter.ttk import Separator


sep = [",", ";", "\t", "|", " "] # possible separators

try:
    for file in f:
        # get index of file in f
        i = f.index(file)
        if i == 0:
            # read first file
            df = pd.read_csv('collision_data/' + file, error_bad_lines=False, encoding = "latin1")
        else:
            # concatenate other files
            # omite errors reading csv
            #df = pd.concat([df, pd.read_csv('collision_data/' + file, error_bad_lines=False)], ignore_index=True)
            df = pd.concat([df, pd.read_csv('collision_data/' + file, error_bad_lines=False, encoding = "latin1")], ignore_index=True)
except:
    # replace byte 0xe9 with 0x20
    df = pd.read_csv('collision_data/' + file, error_bad_lines=False, encoding = "latin1")
    print('Error reading file: ' + file)
```

```python
# count number of elements in df with nan
id = df[df["Num_Acc"].notnull()]

df_col = df[df["col"].notnull()]
```

```python
df_col
```

```python
# get df["col"] where the value is 1
df_frontal = df_col[df_col["col"] == 1]

# get df["col"] where the value is 2
df_rear = df_col[df_col["col"] == 2]
```

```python
# group dataframe by year-> an
df_frontal_y = df_frontal.groupby("an").count()
df_rear_y = df_rear.groupby("an").count()
```

```python
# get filenames with walk
f = []
for (dirpath, dirnames, filenames) in walk('vehicle_data/'):
    f.extend(filenames)
    break
```

```python
try:
    for file in f:
        # get index of file in f
        i = f.index(file)
        if i == 0:
            # read first file
            df_ve = pd.read_csv('vehicle_data/' + file, error_bad_lines=False, encoding = "latin1")
        else:
            # concatenate other files
            # omite errors reading csv
            #df = pd.concat([df, pd.read_csv('collision_data/' + file, error_bad_lines=False)], ignore_index=True)
            df_ve = pd.concat([df, pd.read_csv('vehicle_data/' + file, error_bad_lines=False, encoding = "latin1")], ignore_index=True)
except:
    # replace byte 0xe9 with 0x20
    df_ve = pd.read_csv('vehicle_data/' + file, error_bad_lines=False, encoding = "latin1")
    print('Error reading file: ' + file)
```

```python
# inner join df_ve and df_col
df_ve_col = pd.merge(df_ve, df_col, on='Num_Acc')


```

```python
# group dataframe by an and col
df_ve_col_y = df_ve_col.groupby(["an_y","catv"])
```

```python
df_ve_col_y["catv"]
```

```python
# filter df_ve_col_y by catv == 7
df_ve_col_y_vl  = df_ve_col[df_ve_col["catv"] == 7] # 7 = voiture légère

# filter df_ve_col_y by catv == 8
df_ve_col_y_vl2 = df_ve_col[df_ve_col["catv"] == 8] # 8 = voiture légère + caravana

# filter df_ve_col_y by catv == 9
df_ve_col_y_vl3 = df_ve_col[df_ve_col["catv"] == 9] # 9 = voiture légère + remorque

# filter df_ve_col_y by catv == 10
df_ve_col_y_vl4 = df_ve_col[df_ve_col["catv"] == 10] # 10 = voiture 1.5T-3.5T]
```

```python
# plot df_ve_col_y_vl Num_Acc grouped by col_y 
df_ve_col_y_vl.groupby("col_y").count()["Num_Acc"].plot(kind="bar", title="Nombre de collision par type de collision")
```

```python

# plot df_ve_col_y_vl4 Num_Acc grouped by col_y 
df_ve_col_y_vl4.groupby("col_y").count()["Num_Acc"].plot(kind="bar", title="Nombre de collision par type de collision")
```

```python
# plot histogram df_ve_col_y by an_y
df_ve_col_y_vl["Num_Acc"].unstack().plot(kind='bar', stacked=True, figsize=(15, 10))
# set title
plt.title("Number of collisions by year and type of collision")
#set legend
#plt.legend(["Collision with a fixed object", "Collision with a vehicle in the same direction", "Collision with a vehicle in the opposite direction"])
```
