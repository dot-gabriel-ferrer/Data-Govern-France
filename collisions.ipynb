```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from os import walk
```

```python
# get filenames with walk
f = []
for (dirpath, dirnames, filenames) in walk('collision_data/'):
    f.extend(filenames)
    break
```

```python
f
```

```python
# read dataframe in csv with pandas
df = []
for file in f:
    df.append(pd.read_csv('collision_data/' + file))
```

```python
# join dataframes
df = pd.concat(df)
```
